```python















# ⚙️ 1. INSTALL (run once per runtime)
!pip install -q deep_sort_realtime opencv-python-headless

# ⚙️ 2. IMPORTS
import cv2, csv, numpy as np
from ultralytics import YOLO
from deep_sort_realtime.deepsort_tracker import DeepSort

# ⚙️ 3. LOAD MODEL
model = YOLO('/content/drive/MyDrive/reid_project/best.pt')  # ← path to your weights

# ⚙️ 4. INITIALISE TRACKER (tuned)
tracker = DeepSort(
    max_age=20,          # keep “lost” IDs 20 frames
    n_init=2,            # need only 2 hits to confirm a track
    nms_max_overlap=0.6, # stricter duplicate-box rejection
    max_cosine_distance=0.25,   # harder to re-use an ID
)

# ⚙️ 5. VIDEO I/O PATHS
SRC       = '/content/drive/MyDrive/reid_project/v3.mp4'
OUT_VIDEO = '/content/v3_annotated_final.mp4'
CSV_PATH  = '/content/v3_tracks_final.csv'

# ⚙️ 6. PROCESS
cap    = cv2.VideoCapture(SRC)
w, h   = int(cap.get(3)), int(cap.get(4))
fps    = cap.get(cv2.CAP_PROP_FPS)
writer = cv2.VideoWriter(OUT_VIDEO, cv2.VideoWriter_fourcc(*'mp4v'), fps, (w, h))

with open(CSV_PATH, 'w', newline='') as fcsv:
    csvw = csv.writer(fcsv)
    csvw.writerow(['frame', 'track_id', 'x1', 'y1', 'x2', 'y2', 'conf'])

    frame_idx = 0
    while cap.isOpened():
        ok, frame = cap.read()
        if not ok:
            break

        # ─── YOLO DETECTION ───────────────────────────────────────────
        yolo_res = model(frame, imgsz=960, conf=0.25, verbose=False)[0]

        dets = []  # list of ([x1,y1,x2,y2], conf, class_id)
        for xyxy, conf, cls in zip(yolo_res.boxes.xyxy.cpu().tolist(),
                                   yolo_res.boxes.conf.cpu().tolist(),
                                   yolo_res.boxes.cls.cpu().tolist()):
            if int(cls) != 2 or conf < 0.25:        # keep **players** only
                continue

            x1, y1, x2, y2 = map(int, xyxy)


            shrink = 0.40  # 40% tighter boxes = tighter torso-only fit
            cx, cy = (x1 + x2) // 2, (y1 + y2) // 2
            bw, bh = x2 - x1, y2 - y1
            nw, nh = int(bw * (1 - shrink)), int(bh * (1 - shrink))
            x1 = max(0, cx - nw // 2); y1 = max(0, cy - nh // 2)
            x2 = min(w, cx + nw // 2); y2 = min(h, cy + nh // 2)
            # # shrink box 15 % for visual clarity
            # cx, cy = (x1 + x2) // 2, (y1 + y2) // 2
            # bw, bh = x2 - x1, y2 - y1
            # shrink = 0.15
            # nw, nh = int(bw * (1 - shrink)), int(bh * (1 - shrink))
            # x1 = max(0, cx - nw // 2); y1 = max(0, cy - nh // 2)
            # x2 = min(w, cx + nw // 2); y2 = min(h, cy + nh // 2)

            dets.append(([x1, y1, x2, y2], float(conf), 2))

        # ─── TRACKING ────────────────────────────────────────────────
        tracks = tracker.update_tracks(dets, frame=frame)

        for t in tracks:
            if not t.is_confirmed():
                continue

            tid          = t.track_id
            x1, y1, x2, y2 = map(int, t.to_ltrb())
            conf         = t.det_conf if t.det_conf else 0.0

            csvw.writerow([frame_idx, tid, x1, y1, x2, y2, conf])

            cv2.rectangle(frame, (x1, y1), (x2, y2), (0,255,0), 2)
            cv2.putText(frame, f'ID:{tid}', (x1, y1-8),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0,255,0), 2)

        writer.write(frame)
        frame_idx += 1

cap.release(); writer.release()
print('✅ Done!  ➡', OUT_VIDEO, CSV_PATH)




```
